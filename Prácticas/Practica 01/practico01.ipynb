#Importamos el dataset

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import csv

pesos = []
estaturas = []

with open('datos_peso_estatura.csv', mode='r') as file:
    reader = csv.reader(file)
    next(reader)  
    for row in reader:
        peso, estatura = float(row[0]), float(row[1])
        pesos.append(peso)
        estaturas.append(estatura)

pesos = np.array(pesos)
estaturas = np.array(estaturas)
```

#Graficar los datos

```python
plt.scatter(estaturas, pesos, color='blue', label='Datos generados')
plt.title('Relaci√≥n entre Peso y Estatura')
plt.xlabel('Estatura (m)')
plt.ylabel('Peso (kg)')
plt.legend()
plt.show()
```

#encontramos el modelo y ajustamos

```python
def modelo(x, a, b):
    return a * x + b

# Ajustar la curva a los datos
popt, pcov = curve_fit(modelo, estaturas, pesos)
```

#Graficar el ajuste

```python
plt.scatter(estaturas, pesos, color='blue', label='Datos generados')
plt.plot(estaturas, modelo(estaturas, *popt), color='red', label='Curva ajustada')
plt.title('Ajuste de la curva a la relaci√≥n entre Peso y Estatura')
plt.xlabel('Estatura (m)')
plt.ylabel('Peso (kg)')
plt.legend()
plt.show()

print(f"Par√°metros de la curva ajustada: a = {popt[0]:.2f}, b = {popt[1]:.2f}")
```
